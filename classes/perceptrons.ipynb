## Historical context (short)

- The idea of simple threshold neurons goes back to McCulloch & Pitts (1943) (the McCulloch–Pitts neuron). This is the biological-inspired threshold logic unit. :contentReference[oaicite:2]{index=2}
- Hebb (1949) proposed the idea that co-active neurons strengthen connections — an early learning rule inspiration (Hebb's rule). :contentReference[oaicite:3]{index=3}
- Rosenblatt introduced the Perceptron (late 1950s) — the first artificial neural network with an explicit learning rule. Minsky & Papert (1969) famously showed limitations (e.g., XOR) which triggered an "AI winter" for perceptrons. :contentReference[oaicite:4]{index=4} :contentReference[oaicite:5]{index=5}


## McCulloch–Pitts neuron (threshold logic)

- Model: sum (weights × inputs) then apply a threshold. Output is binary (0/1) or bipolar (-1/+1).
- For binary inputs x1,x2 and threshold θ:
  output = 1 if w1·x1 + w2·x2 ≥ θ, else 0.
  -
<img src= "images/img.1.png"  width="300">

- This simple unit can implement logic gates like AND, OR, NOT by choosing weights and θ appropriately. Example solutions from the slides:
  - OR -> w1=2, w2=2, θ=2 (one valid solution).
  - AND -> w1=1, w2=1, θ=2 (one valid solution).
  - NOT -> w1 = -1 (with appropriate bias/threshold).

(These are analytic examples showing that linearly separable boolean functions can be implemented by a single threshold unit.)


## 1) Intuition & Geometry

- A **perceptron** is the simplest linear binary classifier. It computes a linear score and applies a step activation:

  score = wᵀ x + b

  ŷ = +1 if score ≥ 0, else -1

- The **decision boundary** is the hyperplane wᵀ x + b = 0. In 2D this is a line. The weight vector **w** is perpendicular to the hyperplane and points toward the side predicted +1.

- **Linear separability:** If classes can be separated by a straight line (hyperplane), the perceptron can find a separating hyperplane (under assumptions).

**Intuition for updates:** For a misclassified point (xᵢ, yᵢ ∈ {+1,-1}), update w ← w + η yᵢ xᵢ. This increases yᵢ (wᵀ xᵢ) and nudges the classifier toward correct classification.

**Activation note:** Classic perceptron uses a step function (non-differentiable). That's one reason people prefer logistic regression or neural nets (differentiable).
